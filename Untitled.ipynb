## Data analysis

```python
cars_data['name'].unique()
```

```python
def get_brand_name(car_name):
    car_name = car_name.split(' ')[0]
    return car_name.strip(' ')
```

```python
cars_data['name'] = cars_data['name'].apply(get_brand_name)
cars_data
```

```python
cars_data['mileage'] = cars_data['mileage'].apply(get_brand_name)
```

```python
cars_data['engine'] = cars_data['engine'].apply(get_brand_name)
```

```python
cars_data['max_power'] = cars_data['max_power'].apply(get_brand_name)
```

```python
cars_data.iloc[4933]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
cars_data['name']=encoder.fit_transform(cars_data['name'])
```

```python
cars_data['fuel']=encoder.fit_transform(cars_data['fuel'])
```

```python
cars_data['seller_type']=encoder.fit_transform(cars_data['seller_type'])
```

```python
cars_data['transmission']=encoder.fit_transform(cars_data['transmission'])
```

```python
cars_data['owner']=encoder.fit_transform(cars_data['owner'])
```

```python
cars_data.reset_index(inplace=True)
```

```python
cars_data.drop(columns='index',inplace=True)
```

## Linear Regression

```python
X_train,X_test,y_train,y_test=train_test_split(input,target,test_size=0.2,random_state=42)
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
